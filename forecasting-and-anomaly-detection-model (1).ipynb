```python
# Menemukan model yang baik untuk pengaplikasian forecasting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, recall_score, mean_squared_error, mean_absolute_percentage_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Membaca data dari file Excel
df = pd.read_excel(r'C:/Users/rifal/df_fixed.xlsx')

# Data preprocessing
df['DateTime'] = pd.to_datetime(df['DateTime'])  # Mengubah kolom DateTime menjadi tipe datetime
df.set_index('DateTime', inplace=True)  # Menetapkan kolom DateTime sebagai indeks
dataset = df["Conductivity"]  # Mengambil kolom Conductivity
dataset = pd.DataFrame(dataset)
data = dataset.values  # Mengonversi dataset menjadi array numpy

# Scaling data (feature engineering)
scaler = StandardScaler()  # Menggunakan StandardScaler
scaled_data = scaler.fit_transform(data)  # Melakukan scaling pada data

# Splitting data train dan test
train_size = int(len(data) * 0.75)  # Menentukan ukuran data train
test_size = len(data) - train_size  # Menentukan ukuran data test
train_data = scaled_data[0:train_size, :]  # Membagi data menjadi data train
test_data = scaled_data[train_size:len(data), :]  # Membagi data menjadi data test

# Membuat training dan test sets dengan time-steps
time_steps = 60  # Menetapkan jumlah time steps (langkah waktu) yang digunakan dalam pembentukan data input untuk model LSTM
x_train, y_train = [], []  # Inisialisasi list kosong untuk data pelatihan input (x_train) dan label (y_train)

# Loop untuk membuat data pelatihan
for i in range(time_steps, len(train_data)):
    x_train.append(train_data[i-time_steps:i, 0])  # Menambahkan sekuens data dari train_data ke x_train
    y_train.append(train_data[i, 0])  # Menambahkan nilai target yang sesuai ke y_train
x_train, y_train = np.array(x_train), np.array(y_train)  # Mengubah list menjadi array numpy

x_test, y_test = [], []  # Inisialisasi list kosong untuk data pengujian input (x_test) dan label (y_test)

# Loop untuk membuat data pengujian
for i in range(time_steps, len(test_data)):
    x_test.append(test_data[i-time_steps:i, 0])  # Menambahkan sekuens data dari test_data ke x_test
    y_test.append(test_data[i, 0])  # Menambahkan nilai target yang sesuai ke y_test
x_test, y_test = np.array(x_test), np.array(y_test)  # Mengubah list menjadi array numpy

# Mengubah bentuk data menjadi 3D untuk LSTM [samples, time-steps, features]
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))  # Mengubah bentuk x_train untuk memenuhi input LSTM
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))  # Mengubah bentuk x_test untuk memenuhi input LSTM

# Build LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(50, activation='relu', return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(x_train, y_train, epochs=6, batch_size=32, validation_data=(x_test, y_test))

# Melakukan prediksi menggunakan model yang telah dilatih
y_pred = model.predict(x_test)

# Mengembalikan skala prediksi ke skala asli menggunakan invers transformasi
y_pred_rescaled = scaler.inverse_transform(y_pred)

# Mengembalikan skala nilai uji ke skala asli menggunakan invers transformasi
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Membinarisasi prediksi dan nilai aktual untuk metrik klasifikasi
# Menentukan ambang batas (threshold) sebagai median dari nilai uji yang telah diskalakan kembali
threshold = np.median(y_test_rescaled)

# Mengubah nilai uji asli menjadi biner berdasarkan ambang batas
y_test_binary = (y_test_rescaled > threshold).astype(int)

# Mengubah prediksi menjadi biner berdasarkan ambang batas
y_pred_binary = (y_pred_rescaled > threshold).astype(int)

# Menghitung akurasi klasifikasi
accuracy = accuracy_score(y_test_binary, y_pred_binary)

# Menghitung skor F1 dengan rata-rata makro
f1 = f1_score(y_test_binary, y_pred_binary, average='macro')

# Menghitung recall dengan rata-rata makro
recall = recall_score(y_test_binary, y_pred_binary, average='macro')

# Menghitung Root Mean Squared Error (RMSE) dan membulatkannya ke dua desimal
rmse = np.sqrt(mean_squared_error(y_test, y_pred)).round(2)

# Menghitung Mean Squared Error (MSE)
mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)

# Menghitung Mean Absolute Percentage Error (MAPE)
mape = mean_absolute_percentage_error(y_test_rescaled, y_pred_rescaled)

print(f'Accuracy: {accuracy}')
print(f'F1 Score: {f1}')
print(f'Recall: {recall}')
print(f'RMSE: {rmse}')
print(f'MSE: {mse}')
print(f'MAPE: {mape}')

# Visualization: Loss saat melakukan training
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss during Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Visualization: Actual vs Prediction
plt.figure(figsize=(12, 6))
plt.plot(y_test_rescaled, label='Actual')
plt.plot(y_pred_rescaled, label='Prediction')
plt.title('Actual vs Prediction')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Visualization: Training, Test, dan Prediction
train_indices = range(time_steps, train_size)
test_indices = range(train_size, train_size + len(y_test_rescaled))

plt.figure(figsize=(14, 7))
plt.plot(train_indices, scaler.inverse_transform(train_data[time_steps:train_size]), label='Training Data')
plt.plot(test_indices, y_test_rescaled, label='Test Data')
plt.plot(test_indices, y_pred_rescaled, label='Predictions')
plt.title('Training, Test, and Predictions')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Analisis Error
# Menghitung error absolut antara nilai aktual dan prediksi
errors = np.abs(y_test_rescaled - y_pred_rescaled)

# Mengurutkan indeks berdasarkan nilai error dari kecil ke besar
sorted_indices = np.argsort(errors.flatten())

# Mengambil 5 indeks dengan error terkecil (prediksi yang benar)
correct_indices = sorted_indices[:5]

# Mengambil 5 indeks dengan error terbesar (prediksi yang salah)
incorrect_indices = sorted_indices[-5:]

# Mencetak 5 prediksi dengan error terkecil
print("Prediksi yang Benar (5 instance):")
for i in correct_indices:
    # Mencetak nilai aktual dan prediksi untuk setiap instance yang benar
    print(f"Actual: {y_test_rescaled[i][0]}, Prediction: {y_pred_rescaled[i][0]}")

# Mencetak 5 prediksi dengan error terbesar
print("\nPrediksi yang Salah (5 instance):")
for i in incorrect_indices:
    # Mencetak nilai aktual dan prediksi untuk setiap instance yang salah
    print(f"Actual: {y_test_rescaled[i][0]}, Prediction: {y_pred_rescaled[i][0]}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Konversi kolom 'DateTime' menjadi tipe data datetime
df = pd.read_excel(r'C:/Users/rifal/df_fixed.xlsx')
df['DateTime'] = pd.to_datetime(df['DateTime'])

def train_and_visualize(df, feature):
    # Variabel untuk pelatihan
    cols = [feature]  # Gunakan hanya kolom yang ditentukan untuk peramalan

    # Dataframe baru dengan hanya data pelatihan - 'DateTime' dan kolom fitur yang ditentukan
    df_for_training = df[cols].astype(float)

    # Feature engineering
    scaler = StandardScaler()
    scaler = scaler.fit(df_for_training)
    df_for_training_scaled = scaler.transform(df_for_training)

    # Jumlah waktu yang akan datang
    n_future = 1
    # Jumlah waktu sebelumnya untuk memprediksi waktu depan.
    n_past = 14  

    # Daftar kosong yang akan diisi dengan data pelatihan yang diformat
    trainX = []
    trainY = []

    # Format ulang data input ke dalam bentuk: (n_samples x timesteps x n_features)
    for i in range(n_past, len(df_for_training_scaled) - n_future + 1):
        trainX.append(df_for_training_scaled[i - n_past:i, 0])
        trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

    trainX, trainY = np.array(trainX), np.array(trainY)

    # Bentuk ulang trainX untuk menambahkan dimensi tambahan untuk input LSTM
    trainX = trainX.reshape((trainX.shape[0], trainX.shape[1], 1))

    # Definisikan model LSTM
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
    model.add(LSTM(50, activation='relu', return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(trainY.shape[1]))

    model.compile(optimizer='adam', loss='mse')
    model.summary()

    # Fit model
    history = model.fit(trainX, trainY, epochs=6, batch_size=16, validation_split=0.1, verbose=1)

    plt.plot(history.history['loss'], label='Training loss')
    plt.plot(history.history['val_loss'], label='Validation loss')
    plt.legend()
    plt.title(f'{feature} - Loss during Training')
    plt.show()

    # Memprediksi...
    # Prediksi sekian s setelahnya
    n_days_for_prediction = 30

    predict_period_dates = pd.date_range(df['DateTime'].iloc[-n_past], periods=n_days_for_prediction, freq='S')

    # Membuat prediksi
    prediction = model.predict(trainX[-n_days_for_prediction:])
    y_pred_future = scaler.inverse_transform(prediction)[:, 0]

    # Plot peramalan untuk seluruh dataset
    plt.figure(figsize=(10,6))
    plt.plot(df['DateTime'], df[feature], label='Actual')
    plt.plot(predict_period_dates, y_pred_future, label='Forecast')
    plt.xlabel('Time')
    plt.ylabel(feature)
    plt.legend()
    plt.title(f'{feature} - Actual vs Forecast (Full)')
    plt.show()

    # Plot peramalan yang diperbesar pada bagian akhir dataset
    plt.figure(figsize=(10,6))
    plt.plot(df['DateTime'], df[feature], label='Actual')
    plt.plot(predict_period_dates, y_pred_future, label='Forecast')
    plt.xlabel('Time')
    plt.ylabel(feature)
    plt.legend()
    plt.title(f'{feature} - Actual vs Forecast (Zoomed)')
    
    # Memperbesar bagian akhir dataset dan peramalan
    plt.xlim(df['DateTime'].iloc[-n_past], predict_period_dates[-1])
    
    plt.show()

# Latih dan visualisasikan untuk setiap fitur
features = ['Conductivity', 'Pressure', 'WaterLevel']
for feature in features:
    train_and_visualize(df, feature)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Membaca data dari file Excel
df = pd.read_excel(r'C:/Users/rifal/df_fixed.xlsx')

# Inisialisasi Isolation Forest dengan parameter kontaminasi
clf = IsolationForest(contamination=0.00211, random_state=42)

# Menentukan fitur yang akan digunakan
features = ['Conductivity', 'Pressure', 'WaterLevel']

# Melatih model pada setiap fitur dan mendeteksi anomali
for feature in features:
    X = df[[feature]]  # Memilih fitur yang akan dianalisis
    clf.fit(X)  # Melatih model Isolation Forest pada fitur yang dipilih
    df[f'anomaly_{feature}'] = clf.predict(X)  # Menyimpan hasil prediksi anomali ke dalam DataFrame

# Menampilkan data anomali untuk masing-masing fitur
for feature in features:
    anomalies = df[df[f'anomaly_{feature}'] == -1]  # Memilih baris yang terdeteksi sebagai anomali
    print(f"Data dengan anomali pada fitur {feature}:")
    print(anomalies)  # Mencetak data yang mengandung anomali

# Visualisasi hasil deteksi anomali untuk setiap fitur
plt.figure(figsize=(15, 10))  # Membuat figur dengan ukuran 15x10 inci

for i, feature in enumerate(features, 1):
    plt.subplot(3, 1, i)  # Membuat subplot untuk setiap fitur
    plt.plot(df.index, df[feature], label='Data Asli')  # Memplot data asli
    anomalies = df[df[f'anomaly_{feature}'] == -1]  # Memilih data yang terdeteksi sebagai anomali
    plt.scatter(anomalies.index, anomalies[feature], color='red', label='Anomali')  # Memplot anomali sebagai titik merah
    plt.fill_between(df.index, min(df[feature]), max(df[feature]), where=df[f'anomaly_{feature}'] == -1, color='red', alpha=0.3)  # Menambahkan shading merah untuk area anomali
    plt.title(f'Deteksi Anomali pada Fitur {feature} menggunakan Isolation Forest')  # Menambahkan judul untuk setiap subplot
    plt.xlabel('Index Data')  # Menambahkan label sumbu x
    plt.ylabel(f'Nilai {feature}')  # Menambahkan label sumbu y
    plt.legend() 

plt.tight_layout()  # Mengatur layout agar tidak saling tumpang tindih
plt.show()  # Menampilkan plot
```
